```python
"""
Tables structure
"""

from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy import Column, Integer, Float, String, Date
from sqlalchemy import ForeignKey

Base = declarative_base()

class AirBNB(Base):
    __tablename__ = "airBNB"
    id = Column(Integer, primary_key=True)
    name = Column(String(50))
    host_id = Column(Integer, ForeignKey("host.id"))
    location_id = Column(Integer, ForeignKey("location.id"))
    room_type = Column(String(50))
    price = Column(Integer)
    minimum_nights = Column(Integer)
    review_id = Column(Integer, ForeignKey("review.id"))
    availability_365 = Column(Integer)
    
class Host(Base):
    __tablename__ = "host"
    id = Column(Integer, primary_key=True)
    name = Column(String(50))
    number_of_properties = Column(Integer)
    
    def __repr__(self):
        return "<Host (id='{0}', name={1}, number_of_properties={2})>".format(
               self.id, self.name, self.number_of_properties)

class Location(Base):
    __tablename__ = "location"
    id = Column(Integer, primary_key=True)
    latitude = Column(Float)
    longitude = Column(Float)
    district_id = Column(Integer, ForeignKey("district.id"))
    
    def __repr__(self):
        return "<Location (id='{0}', latitude={1}, longitude={2}, district_id={3})>".format(
               self.id, self.latitude, self.longitude, self.district_id)

class District(Base):
    __tablename__ = "district"
    id = Column(Integer, primary_key=True)
    district = Column(String(50))
    neighbourhood = Column(String(50))

    def __repr__(self):
        return "<District (id='{0}', district={1}, neighbourhood={2})>".format(
               self.id, self.district, self.neightbourhood)

class Review(Base):
    __tablename__ = "review"
    id = Column(Integer, primary_key=True)
    number_of_reviews = Column(Integer)
    last_review = Column(Date)
    
    def __repr__(self):
        return "<Review (id='{0}', number_of_reviews={1}, last_review={2})>".format(
               self.id, self.number_of_reviews, self.last_review)

#Base.metadata.create_all(engine)
```
